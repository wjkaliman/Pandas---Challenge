```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
purchase_data = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(purchase_data)
purchase_data_df .head()
```

## Player Count

* Display the total number of players


```python
# need total count- using lo to pull out rows and columns- drop duplicates
total_df = purchase_data_df.loc[:,["SN", "Age"]].drop_duplicates()

#setting variable of total number players to count
number_of_players = len(total_df)

len(total_df)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# need total unique items, average price, total of purchases and total revenue

items_count = purchase_data_df["Item ID"].unique()
items_count = len(items_count)

average_price_df = purchase_data_df["Price"].mean()
average_price_df
total_purchases_df = len(purchase_data_df)
total_purchases_df
total_revenue_df = purchase_data_df["Price"].sum()
total_revenue_df.round(2)

# Need to convert to DataFrame- done

purchasing_data_df = pd.DataFrame({"Number of Unique Items":items_count,
                                             "Average Purchase Price":[average_price_df],
                                            "Total Number of Purchases":[total_purchases_df],
                                             "Total Revenue":[total_revenue_df]})
#need to format to $- done
purchasing_data_df["Average Purchase Price"] = purchasing_data_df["Average Purchase Price"].map("${:,.2f}".format)
purchasing_data_df["Total Revenue"] = purchasing_data_df["Total Revenue"].map("${:,.2f}".format)
purchasing_data_df

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# need total males, females and ND count and %

gender_count_df = purchase_data_df.groupby("Gender")["SN"].nunique()
gender_count_df.head
#number of players is from above cell-done
gender_percentage_df = gender_count_df/number_of_players
gender_percentage_df.round(2)



# Need to convert to DataFrame-done

gender_demographics_df = pd.DataFrame({"Gender Count": gender_count_df,"Gender Percentage":gender_percentage_df})
# map for format %- done
gender_demographics_df["Gender Percentage"] = gender_percentage_df.map("{:,.2%}".format)

gender_demographics_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Need average purchase price and purchase total by gender

gender_purchases_df = purchase_data_df.groupby("Gender")["Price"].count()


gender_average_df = purchase_data_df.groupby("Gender")["Price"].mean()
gender_average_df.round(2)
gender_total_df = purchase_data_df.groupby("Gender")["Price"].sum()
gender_total_df


gender_total_df = gender_total_df/gender_count_df
gender_total_df.round(2)

# Need to convert to DataFrame-done

gender_analysis_df = pd.DataFrame({"Purchase Count":gender_purchases_df, 
                                   "Average Purchase Price":gender_average_df,
                                   "Total Purchase Value":gender_total_df,
                                  "Average purchase total per person by Gender": gender_total_df})
# .map format for dollar conversion
gender_analysis_df["Average Purchase Price"] = gender_analysis_df["Average Purchase Price"].map("${:,.2f}".format)
gender_analysis_df["Total Purchase Value"] = gender_analysis_df["Total Purchase Value"].map("${:,.2f}".format)
gender_analysis_df["Average purchase total per person by Gender"] = gender_analysis_df["Average purchase total per person by Gender"].map("${:,.2f}".format)
gender_analysis_df
                                   
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,10,15,20,25,30,35,40, 45]
age_ranges = ["<10", "10-14","15-19", "20-24", "25-29", "30-34", "35-39", ">=40"]
total_df["Age groups"] = pd.cut(total_df["Age"], bins, labels=age_ranges)

total_df_counts = total_df["Age groups"].value_counts()
total_df_counts
#need to creat a new column for %
total_percent = total_df_counts/number_of_players

#need new df with age groups and total percent
final_count_df = pd.DataFrame({"total":total_df_counts,"total_percent":total_percent})

#sorting index
final_count_df = final_count_df.sort_index()
final_count_df

#map for %--- before the = is what we have. after is what were changing to
final_count_df["total_percent"] = final_count_df["total_percent"].map("{:,.2%}".format)
final_count_df



```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Need overall spending analysis

players_purchase_count_df = purchase_data_df.groupby("SN").count()["Price"].rename("Purchase Count")
players_average_price_df = purchase_data_df.groupby("SN").mean()["Price"].rename("Average Purchase Price")
players_total_df = purchase_data_df.groupby("SN").sum()["Price"].rename("Total Purchase Value")

# Need to Convert to DataFrame-done


total_user_data_df = pd.DataFrame({"Purchase Count":players_purchase_count_df,
                                   "Average Purchase Price": players_average_price_df,
                                   "Total Purchase Value": players_total_df})
# .map format for dollar conversion-done
total_user_data_df["Average Purchase Price"] = total_user_data_df["Average Purchase Price"].map("${:,.2f}".format)
total_user_data_df["Total Purchase Value"] = total_user_data_df["Total Purchase Value"].map("${:,.2f}".format)
total_user_data_df.head()

top_five_spenders = total_user_data_df.sort_values("Total Purchase Value", ascending=False)
top_five_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# need total Items
items_purchase_count_df = purchase_data_df.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")
items_average_price_df = purchase_data_df.groupby(["Item ID", "Item Name"]).mean()["Price"].rename("Average Purchase Price")
items_value_total_df = purchase_data_df.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")

# need to Convert to DataFrame-done


items_purchased_df = pd.DataFrame({"Purchase Count":items_purchase_count_df,
                                   "Item Price":items_average_price_df,
                                   "Total Purchase Value":items_value_total_df,})
# .map format for dollar conversion
items_purchased_df["Item Price"] = items_purchased_df["Item Price"].map("${:,.2f}".format)
items_purchased_df["Total Purchase Value"] = items_purchased_df["Total Purchase Value"].map("${:,.2f}".format)

items_purchased_df

most_popular_items_df = items_purchased_df.sort_values("Purchase Count", ascending=False)
most_popular_items_df.head(5)
```
